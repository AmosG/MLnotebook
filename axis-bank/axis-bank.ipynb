```python
import pandas as pd
df=pd.read_csv('train.csv')

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


df=df.drop(['APP_ID_C','var_l_1','var_l_2','var_l_3','var_l_4','var_l_5','var_l_6','var_l_7','var_l_8'],axis=1)
number=LabelEncoder()
df['var_d5']=number.fit_transform(df['var_d5'].astype('str'))
df['var_d8']=number.fit_transform(df['var_d8'].astype('str'))
df['var_d1']=number.fit_transform(df['var_d1'].astype('str'))
df['var_d7']=number.fit_transform(df['var_d7'].astype('str'))
df['var_d6']=number.fit_transform(df['var_d6'].astype('str'))
df['var_d4']=number.fit_transform(df['var_d4'].astype('str'))
df['var_d3']=number.fit_transform(df['var_d3'].astype('str'))
df['var_d11']=number.fit_transform(df['var_d11'].astype('str'))
df=df.fillna(0)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()
df[['var_d2']] = scaler.fit_transform(df[['var_d2']])
df.head()
```

```python
df['target'].hist()
```

```python
Y=df.iloc[:,0]
X=df.iloc[:,1:]

X.head()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
model = XGBClassifier(eval_metric='auc',min_child_weight=2,max_depth=3,seed=1000)
model.fit(X_train, y_train)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from keras.models import Sequential
from keras.layers import Dense


model2 = Sequential()
model2.add(Dense(5, input_dim=X_train.shape[1], activation='sigmoid'))
model2.add(Dense(5, activation='sigmoid'))
model2.add(Dense(1, activation='sigmoid'))
# Compile model
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model2.fit(X_train, y_train, epochs=15, batch_size=10)
# evaluate the model
scores = model2.evaluate(X_test,y_test)
print("\n%s: %.2f%%" % (model2.metrics_names[1], scores[1]*100))
print("The summary:")
model2.summary()
```
