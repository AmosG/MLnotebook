```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head(10)
```

```python
df=df.drop(['Name','Ticket'],axis=1)
df.head(10)
```

```python
summ=0
count=0

for i in range(df.shape[0]):
    if str(df.iloc[i]['Age'])!='nan':
        summ+=int(df.iloc[i]['Age'])
        count+=1
avg=float(summ/count)
print(avg)
for i in range(df.shape[0]):
    if str(df.iloc[i][4])=='nan':
        df.ix[i,'Age']=avg
df.head(10)
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
#print(df.head())
#df=df.drop(['PassengerId','Cabin'],axis=1)
number=LabelEncoder()
df['Sex']=number.fit_transform(df['Sex'].astype('str'))
df['Embarked']=number.fit_transform(df['Embarked'].astype('str'))
df.head(10)
```

```python
#df=pd.get_dummies(df,['SibSp'],drop_first=True)
df=df.drop(['PassengerId','Cabin'],axis=1)
df1 = pd.get_dummies(df['Pclass'], prefix = 'pclass')
df1.head()
```

```python
df2 = pd.get_dummies(df['SibSp'], prefix = 'sibsp')
df3 = pd.get_dummies(df['Parch'], prefix = 'parch')
df4 = pd.get_dummies(df['Embarked'], prefix = 'embarked')
```

```python
df=df.drop(['Pclass','SibSp','Parch','Embarked'],axis=1)
df.head()
```

```python
df4.head()
```

```python
df=pd.concat([df,df1,df2,df3,df4],axis=1,sort=False)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
np.random.seed(7)
```

```python
X=df.iloc[:,1:]
Y=df.iloc[:,0]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1000)
```

```python
model = Sequential()
model.add(Dense(5, input_dim=X_train.shape[1], activation='sigmoid'))
model.add(Dense(5, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X_train, Y_train, epochs=100, batch_size=10)
# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print("The summary:")
model.summary()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
model = XGBClassifier(eval_metric='auc',min_child_weight=8,max_depth=8,seed=111)
model.fit(X_train, Y_train)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
